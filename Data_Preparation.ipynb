#**Extract Files**

```python
import os
import zipfile

def unzip_folder(zip_file, extract_path):
    # Extract the contents of the zip file
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall(extract_path)

def rename_files(folder_path):
    # Get a list of all files in the folder
    files = [file for file in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, file))]

    # Sort the files alphabetically to ensure sequential renaming
    files.sort()

# Example usage:
zip_file = "/content/drive/MyDrive/K-Fold Purpose/K Fold.zip"
extract_path = "/content/drive/MyDrive/K-Fold Purpose"
rename_files_after_extraction = True

# Unzip the folder
unzip_folder(zip_file, extract_path)

# Rename files after extraction if needed
if rename_files_after_extraction:
    rename_files(extract_path)
```

#**Labeling Files**

```python
import os

def rename_files(folder_path):
    # Get a list of all files in the folder
    files = [file for file in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, file))]

    # Sort the files alphabetically to ensure sequential renaming
    files.sort()

    # Rename each file sequentially
    for i, file in enumerate(files, start=1):
        old_file_path = os.path.join(folder_path, file)
        new_file_name = f"Te_no_{str(i).zfill(4)}{os.path.splitext(file)[1]}"  # Retain the original file extension
        new_file_path = os.path.join(folder_path, new_file_name)
        os.rename(old_file_path, new_file_path)
        print(f"Renamed {old_file_path} to {new_file_path}")

# Example usage:
folder_path = "/content/drive/MyDrive/K-Fold Purpose/Dataset/notumor"
rename_files(folder_path)
```

#**Resize Files**

```python
from PIL import Image
import os

def resize_images(folder_path, target_size=(168, 168)):
    # Create a directory to store resized images
    output_folder = os.path.join(folder_path, "pituitary")
    os.makedirs(output_folder, exist_ok=True)

    # Loop through all files in the folder
    for file_name in os.listdir(folder_path):
        # Check if the file is an image
        if file_name.endswith((".jpg", ".jpeg", ".png", ".bmp")):
            # Open the image file
            image_path = os.path.join(folder_path, file_name)
            with Image.open(image_path) as img:
                # Convert palette (indexed) images to RGB mode
                if img.mode == "P":
                    img = img.convert("RGB")

                # Convert RGBA mode to RGB mode
                if img.mode == "RGBA":
                    img = img.convert("RGB")

                # Resize the image
                resized_img = img.resize(target_size, Image.ANTIALIAS)

                # Save the resized image to the output folder
                output_path = os.path.join(output_folder, file_name)
                resized_img.save(output_path)
                print(f"Resized {file_name} to {target_size}")

# Example usage:
folder_path = "/content/drive/MyDrive/K-Fold Purpose/Dataset/notumor"
resize_images(folder_path)
```
